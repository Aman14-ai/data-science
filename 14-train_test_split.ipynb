```python
import pandas as pd;
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("carpricestts.csv")
df.rename(columns={'Age(yrs)':"Age" , "Sell Price($)":"Price"},inplace=True)
print(df.shape)
df.head()
```

```python
plt.plot(df.Mileage,df.Price,'ro')
```

```python
plt.plot(df.Age,df.Price,'bo')
```

```python
X_train , X_test , y_train, y_test = train_test_split(df.drop("Price",axis=1),df["Price"],test_size=0.2)
print(X_train.shape , X_test.shape , y_train.shape , y_test.shape)
```

```python
lr = LinearRegression();
```

```python
lr.fit(X_train , y_train)
```

```python
y_test
```

```python
y_predict = lr.predict(X_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
round(lr.score(X_test , y_test) * 100,2)
```

```python
lr.intercept_
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
coef_var1 = -0.3525464
coef_var2 = -309.1619098
intercept = 48443.15478494281
```

```python
var1_range = np.linspace(0, 100, 10)
var2_range = np.linspace(0, 1, 10) 
```

```python
X, Y = np.meshgrid(var1_range, var2_range)
Z = intercept + coef_var1 * X + coef_var2 * Y
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.6, rstride=100, cstride=100)
ax.set_xlabel('Independent Variable 1')
ax.set_ylabel('Independent Variable 2')
ax.set_zlabel('Dependent Variable (Target)')
ax.set_title('Linear Regression Plane with Two Predictors')
# ax.legend() # Uncomment if you plot actual data points

plt.show()
```
