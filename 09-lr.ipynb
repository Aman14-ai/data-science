```python
import pandas as pd;
import matplotlib.pyplot as plt;
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df = pd.DataFrame({
    'area': [2600,3000,3200,3600,4000],
    'price': [550000,565000,610000,680000,725000]
})
df
```

```python
plt.plot(df['area'],df['price'],'ro')
plt.show()
```

```python
plt.xlabel("Area in sq ft")
plt.ylabel("Price in US$")
plt.title("Area vs Price")
plt.scatter(df['area'],df['price'],color='red',marker='+')
plt.show()
```

```python
plt.plot(df['price'],df['area'],'bo')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.DataFrame({
    'area':[1000,1500,2300,3540,4120,4560,5490,3460,4750,2300,9000,8600,7100]
})
d
```

```python
d.to_csv("prediction.csv",index=False)
```

```python
plt.xlabel("Area in sq ft")
plt.ylabel("Price in US$")
plt.title("Area vs Price")
plt.scatter(df.area,df.price,color='red',marker='+',label='Actual')
plt.scatter(df.area,reg.predict(df[['area']]),color='blue',label='Predicted')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.legend()
plt.show()
```
