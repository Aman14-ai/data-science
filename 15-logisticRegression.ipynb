```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
%matplotlib inline
```

```python
df = pd.read_csv('insurance_datalogistic.csv')
print(df.shape)
df.head()
```

```python
plt.plot(df.age,df.bought_insurance,'r+')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop("bought_insurance",axis=1),df["bought_insurance"],test_size=0.1)
print(X_train.shape)
X_test
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_test
```

```python
model.predict(X_test)
```

```python
round(model.score(X_test,y_test)*100,2)
```
