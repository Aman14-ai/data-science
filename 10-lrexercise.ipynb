```python
import pandas as pd;
import matplotlib.pyplot as plt;
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df = df.rename(columns={"per capita income (US$)":"pci"})
df.tail(2)
```

```python
plt.plot(df.year,df.pci,'r.')
```

```python
lr = LinearRegression()
lr.fit(df[['year']],df['pci'])
```

```python
lr.predict([[2017]])
```

```python
print(lr.coef_[0])
print(lr.intercept_)
```

```python
828.4650752227104*2017  -1632210.7578554575
```

```python
plt.scatter(df.year,df.pci,color='red',marker='+',label='Actual')
plt.scatter(df.year,lr.predict(df[['year']]),color='blue',marker='+',label='Actual')
plt.plot(df.year,lr.predict(df[['year']]),'b-.',label='Predicted')
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.legend()
plt.show()
```
