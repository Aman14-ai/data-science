```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label="Max Temp")
plt.plot(days,min_t,label="Min Temp")
plt.plot(days,avg_t,"o-.",label="Avg Temp")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather Report")
plt.legend(loc="best",shadow=True,fontsize="large")
plt.grid()
```

## **bar chart**

```python
company = ['google', 'amazon', 'facebook', 'apple', 'microsoft']
revenue = [90, 136, 56, 260, 125]
plt.bar(company, revenue, color=['red', 'blue', 'green', 'orange', 'magenta'])
plt.grid()
```

```python
xpos = np.arange(len(company))
print(xpos)
plt.bar(xpos,revenue, color=['red', 'blue', 'green', 'orange', 'magenta'])
plt.grid()
```

```python

plt.bar(xpos,revenue, color=['red', 'blue', 'green', 'orange', 'magenta'])
plt.xticks(xpos,company)
plt.grid()
```

```python
profit = [10, 20, 30, 40, 50]
```

```python
plt.bar(xpos-0.2,revenue,label='revenue', width=0.4)
plt.bar(xpos+0.2,profit,label="profit", width=0.4)
plt.legend()
plt.xticks(xpos,company)
plt.grid()
```

```python
plt.barh(xpos,revenue,label='revenue')
plt.barh(xpos,profit,label="profit")
plt.legend()
plt.yticks(xpos,company)
plt.grid()
```

## **histogram**

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar,bins=3, rwidth=0.95) 
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150], rwidth=0.95,histtype='step') 
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150], rwidth=0.95,color=['blue','red'],label=['men','women'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150], rwidth=0.95,color=['blue','red'],label=['men','women'],orientation='horizontal')
plt.legend()
```
